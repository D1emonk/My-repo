<h1> Временные ряды

Временной ряд — это последовательность значений некоторого параметра, измеренных в разные моменты времени или за определённые периоды. Каждое значение временного ряда называется уровнем или отсчётом, и для каждого из них обязательно указывается время измерения или порядковый номер. Временные ряды отличаются от обычных выборок данных тем, что при их анализе учитывается взаимосвязь значений с временным параметром, а не только статистические характеристики

<h2> Классификация временных рядов

1. По количеству показателей: одномерные (один параметр) и многомерные (несколько параметров).
- Одномерный временной ряд — это последовательность значений одной переменной, измеряемой в разные моменты времени. Например, ежедневная температура воздуха, часовые цены акций или количество пользователей сайта в минуту. В таком ряде анализируется только одна метрика, и каждое значение зависит, как правило, от предыдущих значений этого же ряда.

  Особенности:
 - Рассматривается только одна переменная.
 - Значение каждого члена ряда может быть представлено как взвешенная сумма предыдущих значений этого же ряда.
 - Используются методы анализа и прогнозирования, такие как скользящее среднее, экспоненциальное сглаживание, модели ARIMA.
- Многомерный временной ряд — это система из нескольких взаимосвязанных одномерных рядов, где значения одной переменной зависят не только от её собственных прошлых значений, но и от значений других переменных, измеряемых одновременно. Например, в экономике можно анализировать одновременно инфляцию, процентные ставки и ВВП, которые влияют друг на друга.

 Особенности:

 - Включает несколько переменных, измеряемых синхронно.

 - Значения одного ряда зависят от предыдущих значений этого и других рядов.

 - Для анализа применяются модели векторной авторегрессии (VAR) и другие многомерные модели.

2. Моментные и интервальные временные ряды

Это два основных типа дискретных временных рядов, которые различаются способом определения значений и интерпретацией данных во времени.

- Моментные временные ряды
 - В моментном временном ряде значения характеризуют состояние исследуемого явления в конкретные моменты времени.

 - Каждый уровень ряда — это значение показателя на определённый момент (например, на начало или конец дня, месяца, года).

 - Особенность: уровни ряда отражают точечные измерения, и их сумма обычно не имеет смысла, так как значения не аккумулируются во времени.

- Интервальные временные ряды
 - В интервальном временном ряде значения отражают величину явления, накопленную или усреднённую за определённый промежуток времени.

 - Каждый уровень ряда — это показатель, рассчитанный за интервал (например, за день, месяц, квартал).

 - Особенность: уровни интервального ряда можно суммировать для получения показателей за более длительные периоды (например, сумма продаж за квартал равна сумме продаж за три месяца).

3. Равноотстоящие и неравноотстоящие временные ряды

Равноотстоящие и неравноотстоящие временные ряды различаются по равенству интервалов времени между последовательными наблюдениями.

- Равноотстоящие временные ряды

 - В таких рядах значения фиксируются через равные промежутки времени.

 - Интервалы между моментами измерения или окончаниями периодов одинаковы и постоянны на протяжении всего ряда.

 - Большинство физических и экономических процессов анализируются именно с помощью равноотстоящих рядов, так как они упрощают математическую обработку и моделирование.

Пример: ежемесячные продажи, ежедневные температуры, почасовой трафик на сайте.

- Неравноотстоящие временные ряды

 - В этих рядах интервалы между наблюдениями не равны и могут быть произвольными.

 - Принцип равенства интервалов не соблюдается, что усложняет анализ и требует специальных методов обработки.

 - Часто встречаются в финансовых данных, например, биржевые индексы, где значения фиксируются только в рабочие дни, а выходные и праздники пропускаются.

Пример: котировки акций, зарегистрированные в разное время торгового дня, или данные о событиях с нерегулярной периодичностью.

4. Полные и неполные временные ряды

Полные и неполные временные ряды отличаются наличием или отсутствием пропущенных значений (отсчётов) в последовательности данных.

- Полные временные ряды

 - В полном временном ряде данные представлены для каждого момента времени или периода без пропусков.

 - Наблюдения идут подряд с равными или заданными интервалами, и отсутствуют пропущенные значения.

 - Такой ряд удобен для анализа, так как не требует дополнительных процедур заполнения пропусков.

Пример: ежедневные замеры температуры воздуха, где данные есть за каждый день без пропусков.

- Неполные временные ряды

 - В неполном временном ряде присутствуют пропуски — отсутствуют данные за некоторые моменты времени или периоды.

 - Пропуски могут возникать из-за сбоев в сборе данных, технических проблем, нерегулярности измерений или других причин.

 - Для анализа неполных рядов часто применяют методы заполнения пропусков: средними значениями, интерполяцией, скользящим средним, использованием предыдущих или последующих значений и др.

Пример: данные о продажах, где отсутствуют показатели за некоторые дни из-за сбоев в учёте.

Почему важно различать

- Анализ и моделирование временных рядов требуют непрерывности данных для корректного выявления трендов, сезонности и других компонентов.

- Пропуски могут искажать результаты и снижать качество прогнозов, поэтому их нужно либо устранять, либо учитывать при построении моделей.

5. Детерминированные и случайные временные ряды

Детерминированные и случайные временные ряды — это два основных типа временных рядов, которые отличаются природой формирования своих значений.

- Детерминированные временные ряды

 - Представляют собой последовательности значений, которые полностью описываются некоторой фиксированной функцией или формулой без случайных компонентов.

 - Такие ряды не содержат случайных колебаний, их поведение полностью предсказуемо.
 - Пример: ряд значений функции yt=2t+5 где каждое значение определяется точной формулой времени t. Другой пример — количество дней в месяцах (31, 28, 31 и т.д.), которое не зависит от случайных факторов.
 - Анализ и прогнозирование детерминированных рядов проще, так как они не требуют учета случайных шумов.

- Случайные (недетерминированные) временные ряды

 - Содержат случайные компоненты, которые нельзя точно предсказать, так как они зависят от случайных факторов, шумов, внешних воздействий.

 - Значения ряда можно рассматривать как реализацию случайной величины, где каждый уровень — результат как систематических (тренд, сезонность), так и случайных влияний.

 - Пример: биржевой курс акций, где на цену влияют множество непредсказуемых факторов; погодные данные с непредсказуемыми колебаниями.

 - Анализ таких рядов требует статистических методов, основанных на изучении средних значений, дисперсии, автокорреляций и других характеристик.

6. Стационарные и нестационарные временные ряды

Стационарные и нестационарные временные ряды различаются по поведению их статистических характеристик во времени.

- Стационарные временные ряды

 - Для стационарного временного ряда среднее значение (математическое ожидание) остаётся постоянным во времени, то есть ряд не имеет тренда.

 - Ковариация между значениями ряда зависит только от расстояния (лага) между ними, а не от конкретного времени наблюдения.

 - Другими словами, распределение и статистические свойства ряда не меняются при сдвиге по времени.

 - Стационарность бывает строгой (полное распределение не меняется) и слабой (постоянство среднего, дисперсии и автоковариации).

 - Пример: ряд рождаемости в России без ярко выраженной сезонности и тренда.

 - Стационарные ряды проще анализировать и моделировать, многие классические методы (AR, MA, ARMA) требуют стационарности данных.

- Нестационарные временные ряды

 - Нестационарный ряд характеризуется изменяющимися во времени статистическими свойствами: среднее, дисперсия или автоковариация меняются.

 - В таких рядах часто присутствуют тренды (возрастающие или убывающие), сезонные колебания или изменение волатильности.

 - Нестационарность усложняет анализ и прогнозирование, так как традиционные модели могут давать неверные результаты.

 - Пример: данные о количестве авиаперелётов, где наблюдается ярко выраженная сезонность и тренд роста пассажиропотока.

 - Существуют два типа нестационарности: детерминированная (например, линейный тренд или сезонность) и стохастическая (случайные процессы, например, случайное блуждание).

<h2> Стационарность

Стационарность — это ключевое свойство временного ряда, означающее, что его статистические характеристики не меняются со временем. В частности, для стационарного ряда:

- Среднее значение остаётся постоянным во времени (отсутствует тренд).

- Дисперсия (вариация значений) не изменяется.

- Ковариация между значениями ряда зависит только от временного сдвига (лага), а не от конкретного времени наблюдения.

Виды стационарности

- Строгая (или сильная) стационарность: распределение вероятностей любого набора наблюдений не меняется при сдвиге по времени. То есть статистика ряда полностью инвариантна относительно времени.

- Слабая (или широкая) стационарность: сохраняются постоянство среднего и ковариации, что достаточно для большинства практических задач анализа временных рядов.

Значение стационарности

Стационарность важна, потому что многие классические методы анализа и прогнозирования временных рядов (например, модели AR, MA, ARMA, ARIMA) требуют, чтобы данные были стационарными. Если ряд нестационарен (имеет тренд, сезонность, изменяющуюся дисперсию), то результаты анализа могут быть некорректными.

Проверка стационарности

- Визуальный анализ графика временного ряда: наличие тренда или изменяющейся дисперсии указывает на нестационарность.

- Анализ автокорреляционной функции.

- Статистические тесты, например:

 - Тест Дики-Фуллера (Augmented Dickey-Fuller, ADF) — проверяет наличие единичного корня, при p-value меньше заданного уровня значимости ряд считается стационарным.

 - Тест KPSS — проверяет гипотезу о стационарности.

Приведение ряда к стационарному виду

Если ряд нестационарен, применяют преобразования:

- Дифференцирование — вычитание предыдущего значения для устранения тренда.

- Удаление тренда (например, с помощью регрессии или фильтра).

- Стабилизация дисперсии (логарифмирование, Box-Cox преобразование).

- Удаление сезонности (например, сезонное дифференцирование).

Проверка стационарности

- Используются статистические тесты, например, расширенный тест Дики-Фуллера (ADF), тест KPSS и тест Филлипса-Перрона.

- Тесты проверяют гипотезы о наличии или отсутствии единичного корня (признак нестационарности).

- На практике результаты тестов могут быть противоречивыми из-за структурных сдвигов и особенностей данных.

<h3> Статистические тесты

1. Тест Дики-Фуллера (Augmented Dickey-Fuller, ADF)

Тест Дики — Фуллера (DF-тест, Dickey — Fuller test) — это методика, которая используется в прикладной статистике и эконометрике для анализа временных рядов для проверки на стационарность.

1.1 Определение нулевой и альтернативной гипотез:

 - H₀ (Нулевая гипотеза): Ряд имеет единичный корень (ρ = 1 или φ = 0). Ряд нестационарен.

 - H₁ (Альтернативная гипотеза): Ряд не имеет единичного корня (ρ < 1 или φ < 0). Ряд стационарен (возможно, вокруг детерминированного тренда/константы).

1.2 Выбор спецификации регрессионной модели:

В зависимости от предполагаемой структуры ряда используются три основные модели:

Модель 1: Без константы и тренда (Random Walk):

 Δyₜ = φyₜ₋₁ + εₜ

 - Δyₜ = yₜ - yₜ₋₁ (первая разность ряда)

 - yₜ₋₁ (лаг первого порядка исходного ряда)

- φ = (ρ - 1) (тестируемый коэффициент)

- εₜ (ошибка, белый шум)

- Используется, если ряд колеблется вокруг нуля без тренда.

Модель 2: С константой, но без тренда (Random Walk with Drift):

Δyₜ = α + φyₜ₋₁ + εₜ

- α (константа или "снос", intercept)

- Используется, если ряд колеблется вокруг ненулевого среднего без тренда.

Модель 3: С константой и линейным трендом (Random Walk with Drift and Trend):

Δyₜ = α + βt + φyₜ₋₁ + εₜ

- β (коэффициент линейного тренда)

- t (временной тренд, обычно t=1,2,3,...,T)

- Используется, если ряд имеет явный линейный тренд.

1.3 Оценивание регрессии методом наименьших квадратов (МНК):

- Для выбранной модели (1, 2 или 3) оцениваются параметры (α, β, φ) с помощью стандартной процедуры МНК.

- Ключевой коэффициент для теста — φ.

1.4 Расчёт t-статистики для коэффициента φ:

- Рассчитывается стандартная t-статистика для коэффициента φ:
t_φ = (φ_hat - φ₀) / SE(φ_hat)

- Важнейший момент: В тесте Дики-Фуллера мы тестируем специфическую нулевую гипотезу H₀: φ = 0 (что эквивалентно ρ = 1). Поэтому формула упрощается:
DF_t = (φ_hat - 0) / SE(φ_hat) = φ_hat / SE(φ_hat)

- Эта t-статистика (DF_t) называется статистикой Дики-Фуллера.

1.5 Сравнение с критическими значениями Дики-Фуллера:

- Критические значения для DF_t НЕ совпадают с критическими значениями стандартного t-распределения Стьюдента! Это связано с нестандартным распределением статистики под нулевой гипотезой (при наличии единичного корня).

- Критические значения были получены Дэвидом Дики и Уэйном Фуллером с помощью метода Монте-Карло и зависят от:

 - Выбранной спецификации модели (1, 2 или 3).

 - Объема выборки (количества наблюдений T).

 - Желаемого уровня значимости (α, обычно 1%, 5%, 10%).

- Таблицы критических значений являются стандартными и встроены во все статистические пакеты (R, Python/statsmodels, Stata, EViews и др.).

1.6 Принятие решения:

- Если DF_t < Критического значения (для выбранного α):

 - Отвергаем нулевую гипотезу H₀ (φ = 0).

 - Вывод: Ряд НЕ имеет единичного корня. Ряд стационарен (в рамках выбранной спецификации модели).

- Если DF_t >= Критического значения (для выбранного α):

 - Не отвергаем нулевую гипотезу H₀ (φ = 0).

 - Вывод: Недостаточно доказательств, чтобы отвергнуть наличие единичного корня. Ряд, вероятно, нестационарен.

2. Тест KPSS

Расчет теста Квятковского-Филлипса-Шмидта-Шина (KPSS) — это процесс проверки стационарности временного ряда, где нулевая гипотеза H₀ гласит, что ряд стационарен (вокруг детерминированного тренда или среднего). Это делает KPSS дополнением к тесту Дики-Фуллера (ADF), где H₀ — нестационарность (единичный корень).

2.1 Определение нулевой и альтернативной гипотез:

- H₀ (Нулевая гипотеза): Временной ряд yₜ стационарен (либо стационарен вокруг уровня, либо стационарен вокруг детерминированного тренда).

- H₁ (Альтернативная гипотеза): Временной ряд yₜ нестационарен (имеет единичный корень).

2.2 Выбор спецификации регрессии (уровень или тренд):

Модель 1: Стационарность вокруг уровня (Constant / Level Stationarity):
yₜ = β₀ + uₜ
Здесь β₀ — константа (уровень), uₜ — стационарный процесс с нулевым средним.

Модель 2: Стационарность вокруг тренда (Trend Stationarity):
yₜ = β₀ + β₁ * t + uₜ
Здесь β₀ — константа, β₁ — коэффициент линейного тренда, t — время, uₜ — стационарный процесс с нулевым средним.

2.3 Оценивание регрессии методом наименьших квадратов (МНК):

Оцените выбранную регрессию (модель 1 или 2) методом МНК. Получите:
- Оценки коэффициентов (β̂₀ и, если применимо, β̂₁).
- Остатки регрессии eₜ:
eₜ = yₜ - ŷₜ
где ŷₜ — предсказанное значение по модели (β̂₀ для модели 1 или β̂₀ + β̂₁ * t для модели 2).

2.4. Расчет частичных сумм остатков:

Постройте последовательность частичных сумм остатков (Partial Sum Process) Sₜ:

Sₜ = ∑ᵗᵢ₌₁ eᵢ для t = 1, 2, ..., T

- где T — общее число наблюдений.

5. Расчет статистики KPSS:

Ключевая статистика теста KPSS вычисляется по формуле:

KPSS = (1 / (T² * s²)) * ∑ₜ₌₁ᵀ Sₜ²

где:
- T — количество наблюдений.
- Sₜ — частичная сумма остатков на момент t (из шага 4).
- s² — оценка долгосрочной (long-run) дисперсии остатков eₜ.

6. Оценка долгосрочной дисперсии (s²):

Это самый важный и сложный шаг в расчете KPSS. Долгосрочная дисперсия s² определяется как:

s² = γ₀ + 2 * ∑ⱼ₌₁^{q} wⱼ * γⱼ

где:

- γⱼ — автоковариация остатков eₜ с лагом j:
γⱼ = (1/T) * ∑ₜ₌ⱼ₊₁ᵀ eₜ * eₜ₋ⱼ
- wⱼ — веса из функции ядра (kernel function). Наиболее часто используют ядро Бартлетта (Bartlett) (как в оригинальной статье KPSS):
wⱼ = 1 - j / (q + 1) для j = 1, 2, ..., q
- q — ширина окна (lag truncation parameter) или "число лагов". Это определяет, сколько автоковариаций учитывать. Выбор q критичен!
- Правило Эндрюса (Andrews, 1991): Автоматически выбирает q на основе данных. Часто используется по умолчанию в статистических пакетах.
- Эвристические правила: Например, q = floor(4*(T/100)^{2/9}) или q = floor(12*(T/100)^{1/4}) (где floor — округление вниз). Суть в том, что q должно расти с ростом T, но медленнее, чем T.

7. Сравнение с критическими значениями:

Критические значения для статистики KPSS зависят от:

- Выбранной модели (только константа / константа и тренд).
- Уровня значимости (α, обычно 1%, 5%, 10%).
- Объема выборки (T), хотя для больших T таблицы сходятся.
- Критические значения для KPSS значительно выше, чем для стандартных тестов (в отличие от ADF, где они ниже).
- Стандартные таблицы критических значений (например, Kwiatkowski et al., 1992) встроены в статистические пакеты.

8. Принятие решения:
* Если KPSS статистика > Критического значения (для выбранного α и модели):
Отвергаем нулевую гипотезу H₀.
Вывод: Ряд НЕ стационарен (вероятно, имеет единичный корень).
* Если KPSS статистика ≤ Критического значения (для выбранного α и модели):
Не отвергаем нулевую гипотезу H₀.
Вывод: Нет достаточных оснований отвергнуть стационарность ряда.

<h3> Основные методы проверки стационарности

1. Анализ среднего

Суть: Проверяется, остаётся ли среднее значение временного ряда постоянным во времени.

- Как проводится:

 - Визуально — строится график ряда, и оценивается наличие тренда (постепенного роста или падения среднего).

 - Статистически — ряд разбивается на несколько равных по длине сегментов, и сравниваются средние значения в этих сегментах. Если средние существенно отличаются — ряд нестационарен.

- Значение: Постоянное среднее — признак стационарности, наличие тренда — признак нестационарности.

2. Анализ дисперсии

Суть: Проверяется, изменяется ли дисперсия (вариация) значений ряда во времени.

- Как проводится:

 - Аналогично среднему, дисперсия оценивается по сегментам ряда.

 - Визуально — наблюдается, меняется ли амплитуда колебаний на графике.

- Значение: Постоянная дисперсия — признак стационарности, изменяющаяся — нестационарности (например, гетероскедастичность).

3. Анализ автокорреляции (ACF) и частичной автокорреляции (PACF)

Суть: Изучается зависимость значений ряда от его предыдущих значений с разными лагами.

- Как проводится:

 - Строятся графики автокорреляционной функции (ACF) и частичной автокорреляционной функции (PACF).

 - Для стационарного ряда автокорреляции быстро убывают и зависят только от лага, а не от времени.

 - Нестационарный ряд часто показывает медленное затухание автокорреляций.

- Значение: Быстро затухающая ACF — признак стационарности, медленно затухающая — нестационарности.

4. Графический анализ

Суть: Визуальный осмотр графика временного ряда и его характеристик.

- Как проводится:

 - Анализируется наличие тренда, сезонности, резких изменений, скачков.

 - Строятся графики ряда, ACF, PACF.

- Значение: Помогает выявить явные признаки нестационарности и выбрать подходящие методы преобразования.

5. Статистические тесты на стационарность

- Тест Дики-Фуллера (ADF):

 - Проверяет наличие единичного корня (признак нестационарности).

 - Нулевая гипотеза: ряд нестационарен.

 - Если p-value < 0.05, гипотеза отвергается — ряд стационарен.

- Тест KPSS:

 - Обратный тест, где нулевая гипотеза — ряд стационарен.

 - Помогает дополнительно подтвердить результаты ADF.

- Тест Филипса-Перрона:

 - Модификация ADF с учётом гетероскедастичности и автокорреляции остатков.

- Как проводится:

 - Рассчитывается статистика теста на основе регрессии разностей ряда.

 - Сравнивается с критическими значениями.

- Значение: Формальная проверка стационарности, обязательна для построения моделей ARIMA и др.

6. Методы сглаживания и выделения тренда

- Скользящее среднее:

 - Усреднение значений ряда за фиксированное окно для выявления тренда.

- Экспоненциальное сглаживание:

 - Придаёт больший вес последним наблюдениям, помогает выделить тренд и сезонность.

- Дифференцирование:

 - Вычитание предыдущего значения ряда для устранения тренда и достижения стационарности.

- Значение: Позволяют преобразовать нестационарный ряд в стационарный для дальнейшего анализа.

7. Спектральный и гармонический анализ

Суть: Анализ частотных компонентов временного ряда для выявления периодичности и циклов.

- Как проводится: Используются преобразования Фурье и вейвлет-анализ.

- Значение: Помогает выявить скрытые циклы, которые влияют на стационарность ряда.

<h3> Модель ARIMA

ARIMA — это расширение модели ARMA, предназначенное для работы с нестационарными временными рядами. Она включает три компонента:

- AR (авторегрессия, порядок p) — текущие значения ряда зависят от нескольких предыдущих значений.

- I (интегрирование, порядок d) — количество раз, которое нужно продифференцировать ряд, чтобы сделать его стационарным.

- MA (скользящее среднее, порядок q) — текущие значения зависят от ошибок (шумов) предыдущих шагов.

Модель обозначается как ARIMA(p, d, q).

Основные шаги работы с ARIMA
1. Проверка стационарности ряда.
Если ряд нестационарен (имеет тренд, меняющуюся дисперсию), применяют дифференцирование — вычитание предыдущих значений, чтобы получить стационарный ряд. Параметр d — число таких дифференцирований.

2. Определение параметров p и q

- p — порядок авторегрессии, определяется по графику частичной автокорреляционной функции (PACF).

- q — порядок скользящего среднего, определяется по графику автокорреляционной функции (ACF).

3. Оценка параметров модели.
Коэффициенты ai и bj оцениваются с помощью методов максимального правдоподобия или наименьших квадратов.

4. Проверка качества модели.
Анализируются остатки модели на предмет отсутствия автокорреляции и нормальности.

5. Прогнозирование.
Используется построенная модель для предсказания будущих значений временного ряда.

Преимущества модели ARIMA

- Позволяет моделировать широкий класс временных рядов, включая нестационарные.

- Учитывает как зависимости от прошлых значений, так и влияние случайных ошибок.

- Гибкая структура, позволяющая адаптироваться под данные.

<h2> Компоненты временных рядов

  Временные ряды обычно рассматривают как сумму нескольких компонентов: тренда, сезонности, цикличности и случайных колебаний. Вот что означают эти термины и как они различаются:

1. Тренд

Тренд — это общая направленная систематическая компонента временного ряда, отражающая долгосрочные изменения данных во времени. Тренд может быть линейным или нелинейным и показывает, растёт ли показатель, снижается или остаётся стабильным на протяжении длительного периода. Например, рост средней глобальной температуры или увеличение продаж компании в течение нескольких лет — это проявления тренда. Тренд отражает устойчивые изменения, отличающиеся от краткосрочных колебаний.

2. Сезонность

Сезонность — это периодически повторяющаяся компонента временного ряда, связанная с календарными или другими регулярными циклами, обычно в пределах одного года. Сезонные колебания проявляются, например, в росте продаж перед новогодними праздниками и их снижении после них. Сезонность характеризуется фиксированным периодом (например, 12 месяцев в году) и повторяемостью в каждом цикле. Для выявления сезонности часто используют автокорреляцию, а в моделях учитывают её через аддитивные или мультипликативные коэффициенты сезонности. Пример — рост спроса на кондиционеры летом и падение зимой.

3. Цикличность

Цикличность — это повторяющиеся, но непериодические колебания временного ряда, которые длятся дольше одного года и не имеют фиксированной длины. Циклы могут иметь разную амплитуду и длительность, отражая, например, экономические подъемы и спады. В отличие от сезонности, циклы не связаны с календарными периодами и зависят от более глобальных факторов, таких как экономическая ситуация. На графике цикличность проявляется как волнообразные колебания вокруг тренда, например, циклы экономической активности с периодами роста и кризисов.

4. Случайные колебания

Случайные колебания — это нерегулярные, непредсказуемые изменения во временном ряду, вызванные случайными факторами, ошибками измерения или внешними воздействиями, не связанными с систематическими компонентами. Они воспринимаются как «шум», который затрудняет выявление тренда, сезонности и цикличности. Методы анализа временных рядов часто включают фильтрацию или сглаживание для уменьшения влияния случайных колебаний.

Таким образом, временной ряд можно представить как сумму:

- Тренда — долгосрочной направленности,

- Сезонности — регулярных годичных колебаний,

- Цикличности — нерегулярных, длительных циклов,

- Случайных колебаний — непредсказуемого шума.

Понимание и выделение этих компонентов позволяет лучше анализировать данные и строить точные прогнозы.
