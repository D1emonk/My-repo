```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/BostonHousing.csv')
df
```

<h3>Зависимость уровня преступности от доступности радиальных шоссе</h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.barplot(x="rad", y="crim", data=df, errorbar=None)

plt.xlabel("Доступность радиальных шоссе (rad)")
plt.ylabel("Уровень преступности (crim)")
plt.title("Зависимость уровня преступности от доступности радиальных шоссе")

plt.show()
```

```python
df[df["crim"] > 10][['crim','rad']]
```

```python
df[df["crim"] < 10][['crim','rad']]
```

Можно заметить, что доля преступности от доступности радиальных шоссе
с коэффициентом 24 пропорционально возрастает, по сравнению с другими
коэффициентами

<h3>Построим корреляционную матрицу признаков</h3>

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Корреляционная матрица признаков")
plt.show()
```

<h3>Зависимость уровня преступности от ставки налога на имущество</h3>

```python
plt.figure(figsize=(6, 4))
sns.barplot(x="tax", y="crim", data=df, errorbar=None)

plt.xlabel("ставка налога на имущество (tax)")
plt.ylabel("Уровень преступности (crim)")
plt.title("Зависимость уровня преступности от ставки налога на имущество")

plt.show()
```

```python
df[df["crim"] > 10][['crim','tax']]
```

```python
df[df["crim"] < 10][['crim','tax']]
```

В данном случаи преступность растет в том случаи, если налог на имущество дорогой. 
Высокие налоги могут указывать на благосостояние района, что связано с преступностью.

<h3>Зависимость уровня преступности от процента малообеспеченного населения</h3>

```python
high_crime = df[df["crim"] > 10]  # Уровень преступности > 10
low_crime = df[df["crim"] <= 10]  # Уровень преступности ≤ 10

plt.figure(figsize=(14, 6))

# График для высокой преступности
plt.subplot(1, 2, 1)
sns.scatterplot(x=high_crime["lstat"], y=high_crime["crim"], color="red")
plt.xlabel("Процент малообеспеченного населения (lstat)")
plt.ylabel("Уровень преступности (crim)")
plt.title("Высокий уровень преступности (crim > 10)")

# График для низкой преступности
plt.subplot(1, 2, 2)
sns.scatterplot(x=low_crime["lstat"], y=low_crime["crim"], color="blue")
plt.xlabel("Процент малообеспеченного населения (lstat)")
plt.ylabel("Уровень преступности (crim)")
plt.title("Низкий уровень преступности (crim ≤ 10)")

# Показываем графики
plt.tight_layout()
plt.show()
```

```python
df[df["crim"] > 10][['crim','lstat']]
```

```python
df[df["crim"] < 10][['crim','lstat']]
```

<h3>Доля малоимущих на рост криминала в дорогих районах</h3>

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["lstat"], y=df["crim"], hue=df["medv"], palette="coolwarm", size=df["medv"], sizes=(10, 200))
plt.xlabel("Доля малоимущих (lstat)")
plt.ylabel("Уровень преступности (crim)")
plt.title("Влияние малоимущего населения на преступность в зависимости от стоимости жилья")
plt.legend(title="Медианная цена жилья")
plt.show()
```

```python
df[df["crim"] > 10][['crim','lstat',"medv"]]
```

```python
df[df["crim"] < 10][['crim','lstat',"medv"]]
```

Проявление роста приступности к проценту малообеспеченного населения данного района проявляется по следующим признакам: 1. Отрицательная корреляция между medv и crim (-0.39). То есть, если район дорогой, то преступность ниже 2. Положительная корреляция между lstat и crim (0.46). Как говорилось в первом пункте, что если район дорогой, то и преступность ниже. Но если в дорогих районах будет увеличиваться малоимущие, то тогда и в таких районах будет увеличиваться преступность.

<h3>Вывод</h3>

Если смотреть Корреляционную матрицу признаков, то можно увидеть, наибольшая положительная корреляция с crim наблюдается 

rad (доступность радиальных шоссе, 0.63) – районы с высокой доступностью к шоссе имеют более высокий уровень преступности.

tax (налог на имущество, 0.58) – в районах с высокими налогами уровень преступности тоже выше.

lstat (процент малообеспеченного населения, 0.46) – чем больше бедного населения, тем выше преступность.

nox (уровень загрязнения воздуха, 0.42) – промышленные районы коррелируют с высокой преступностью.

indus (доля промышленных зон, 0.41) – больше промышленности связано с повышенной преступностью.

age (процент старых домов, 0.35) – в районах со старыми домами преступность выше.

наибольшая отрицательная корреляция с crim наблюдается

medv (стоимость жилья, -0.39) – в дорогих районах уровень преступности ниже.

b (процент чернокожего населения, -0.39) – чем выше этот показатель, тем ниже преступность.

dis (удаленность от рабочих центров, -0.38) – удаленные от города районы менее криминальны.

rm (среднее число комнат, -0.22) – большее число комнат в домах связано с более низкой преступностью.
