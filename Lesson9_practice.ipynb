```python
import matplotlib.pyplot as plt
import seaborn as sns

# Выживаемость по полу
plt.figure(figsize=(6, 4))
sns.barplot(x="Sex", y="Survived", data=df, errorbar=None)
plt.title("Выживаемость по полу")
plt.xlabel("Пол")
plt.ylabel("Доля выживших")
plt.show()
```

```python
# Пассажины мужского пола с возрастом меньше 30 лет 
df[(df["Sex"] == "male") & (df["Age"] < 30)]
```

```python
# Вывод поссажиров, возраст которых неизвестен
df[df["Age"].isna()]
```

```python
# Медианный возраст человека 
df["Age"].median()
```

```python
# к-во людей, которым меньше 18
df[df["Age"] < 18].shape[0]
```

```python
# средний возраст мужчин и женщин
avg = df.groupby("Sex")["Age"].mean()

# Строим столбчатый график
plt.figure(figsize=(5, 4))
sns.barplot(x=avg.index, y=avg.values, hue=avg.index, palette="coolwarm", legend=False)

plt.title("Средний возраст мужчин и женщин")
plt.xlabel("Пол")
plt.ylabel("Средний возраст")
plt.show()
```
